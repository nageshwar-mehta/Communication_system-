```python
import numpy as np 
import matplotlib.pyplot as plt 
```

**original signal**

```python


# Define parameters
dur = 1  # Duration of the signal in seconds
sr = 10000  # Sampling rate in Hz
f0 = 1000  # Frequency of the signal in Hz

# Generate time values
t = np.linspace(-dur, dur, sr * dur * 2)

# Generate complex exponential signal
x_t = np.cos(2*np.pi*f0*t)

# Perform Fourier Transform on the signal
X_f = np.fft.fftshift(np.fft.fft(x_t))

# Generate frequency values for plotting
freq = np.fft.fftshift(np.fft.fftfreq(len(X_f), 1/sr))

# Plot the time-domain signal
plt.figure(figsize=[12, 3])
plt.subplot(1, 2, 1)
plt.title("x(t)")
plt.xlabel("time in sec")
plt.ylabel("Amplitude")
plt.plot(t, x_t)

# Plot the magnitude spectrum of the signal
plt.subplot(1, 2, 2)
plt.title("|X(f)|")
plt.xlabel("frequency in Hz")
plt.ylabel("Amplitude")
plt.xlim(-2000,2000)
plt.plot(freq, abs(X_f / (sr * dur * 2)))
plt.tight_layout()
plt.show()
```

**modulation**

```python


# Define parameters for the modulation
f1 = 200  # Frequency of the modulating signal in Hz
cos_sig = np.cos(2 * np.pi * f1 * t)

# Modulate the unmodulated signal with the cosine signal
x_m_t = x_t * cos_sig

# Perform Fourier Transform on the modulated signal
X_m_f = np.fft.fftshift(np.fft.fft(x_m_t))

# Generate frequency values for plotting
freq = np.fft.fftshift(np.fft.fftfreq(len(X_m_f), 1/sr))

# Plot the modulated signal in the time domain
plt.figure(figsize=[12, 3])
plt.subplot(1, 2, 1)
plt.title("Modulated Signal")
plt.xlabel("Time (in sec)")
plt.ylabel("Amplitude")
plt.plot(t, x_m_t)

# Plot the magnitude spectrum of the modulated signal
plt.subplot(1, 2, 2)
plt.title("FT of Modulated Signal")
plt.xlabel("Frequency (in Hz)")
plt.ylabel("Amplitude")
plt.xlim(-2000,2000)
plt.plot(freq, np.abs(X_m_f) / (sr*dur*2))
plt.tight_layout()
plt.show()
```
